```python
from pathlib import Path
import pandas as pd

# Go one folder above "notebooks"
PROJECT_ROOT = Path.cwd().parent

print("PROJECT_ROOT:", PROJECT_ROOT)

image_df = pd.read_csv(PROJECT_ROOT / "data" / "processed" / "train.csv")
clinical_df = pd.read_csv(PROJECT_ROOT / "data" / "clinical" / "anemia.csv")

print("Image Samples   :", len(image_df))
print("Clinical Samples:", len(clinical_df))

image_df.head()
```

```python
image_df["Geo"] = image_df["country"].map({
    "India": 0,
    "Italy": 1
})

image_df[["country", "Geo"]].head()
```

```python
fusion_df = image_df.copy()

clinical_columns = [
    "Gender",
    "Hemoglobin",
    "MCH",
    "MCHC",
    "MCV"
]

fusion_df[clinical_columns] = clinical_df.loc[
    :len(fusion_df)-1,
    clinical_columns
].values

fusion_df.head()
```

```python
fusion_df = fusion_df[
    [
        "image_path",
        "country",
        "Geo",
        "Gender",
        "Hemoglobin",
        "MCH",
        "MCHC",
        "MCV",
        "label",
        "filename",
        "patient_folder"
    ]
]

fusion_df.head()
```

```python
save_path = PROJECT_ROOT / "data" / "processed" / "fusion_data.csv"

fusion_df.to_csv(save_path, index=False)

print("Fusion Dataset Saved Successfully!")
print(save_path)
```
