```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

```python
import torch

from torch.utils.data import DataLoader

from ml.datasets.geo_dataset import GeoDataset
from ml.models.geo_model import GeoModel

from ml.config import *
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("="*70)
print("Geo Feature Model")
print("="*70)

print("Device :", device)
```

```python
GEO_CSV = PROJECT_ROOT / "data" / "geo" / "geo_risk.csv"

print("Geo CSV :", GEO_CSV)
```

```python
dataset = GeoDataset(GEO_CSV)

print("="*70)
print("Geo Dataset Loaded Successfully")
print("="*70)

print("Total Samples :", len(dataset))
```

```python
sample = dataset[0]

print("="*70)
print("One Sample")
print("="*70)

print(sample)
print(sample.shape)
```

```python
loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True
)

print("="*70)
print("DataLoader Ready")
print("="*70)

print("Total Batches :", len(loader))
```

```python
geo = next(iter(loader))

print("="*70)
print("Batch Loaded")
print("="*70)

print("Batch Shape :", geo.shape)
```

```python
model = GeoModel().to(device)

print("="*70)
print("Geo Model Loaded Successfully")
print("="*70)

print(model)
```

```python
geo = geo.to(device)

model.eval()

with torch.no_grad():

    features = model(geo)

print("="*70)
print("Forward Pass Successful")
print("="*70)

print("Input Shape  :", geo.shape)
print("Output Shape :", features.shape)
```

```python
print("="*70)
print("Extracted Geo Features")
print("="*70)

print(features)
```

```python
print("="*70)
print("Feature Statistics")
print("="*70)

print("Mean :", features.mean().item())
print("Std  :", features.std().item())
print("Min  :", features.min().item())
print("Max  :", features.max().item())
```
