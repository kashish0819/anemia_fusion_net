```python
selected_columns = [

    "District Names",

    "State/UT",

    "Women (age 15-49 years) whose Body Mass Index (BMI) is below normal (BMI <18.5 kg/m2)21 (%)",

    "Children age 6-59 months who are anaemic (<11.0 g/dl)22 (%)",

    "Non-pregnant women age 15-49 years who are anaemic (<12.0 g/dl)22 (%)",

    "All women age 15-49 years who are anaemic22 (%)"

]

geo_selected = geo_df[selected_columns]

geo_selected.head()
```

```python
geo_selected.columns = [

    "District",

    "State",

    "Low_BMI",

    "Child_Anemia",

    "NonPregnant_Women_Anemia",

    "Women_Anemia"

]

geo_selected.head()
```

```python
geo_selected["Geo_Risk"] = (

    0.50 * geo_selected["Women_Anemia"]

    +

    0.30 * geo_selected["Child_Anemia"]

    +

    0.20 * geo_selected["Low_BMI"]

)
```
